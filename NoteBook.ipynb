```python
import urllib
import lxml
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import html5lib
import requests
url_ = "http://www.atlasf1.com/2003/aus/laptimes.html"

page = requests.get(url_)
soup = BeautifulSoup(page.content, 'html5lib')
```

```python
import datetime
table = soup.find("table", class_="copyright")
df = pd.read_html(str(table), flavor='bs4', header=[0])[0]
#df.drop(["Unnamed: 0","Unnamed: 6"],axis=1, inplace=True)
dateparse = lambda x: pd.datetime.strptime(x, '%M:%S.%f')
seconds = lambda x: x.total_seconds()
df["Lap 1"] = df["Lap 1"].apply(dateparse)
df["Lap 1"] = df["Lap 1"] - datetime.datetime(1900,1,1)
df["Lap 1"] = df["Lap 1"].apply(seconds)
df.head()
```
